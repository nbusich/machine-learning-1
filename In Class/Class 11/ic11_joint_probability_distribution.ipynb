```python
import numpy as np
from scipy.special import factorial
import matplotlib.pyplot as plt
a = np.genfromtxt('time_until_phone_drop_exam_4k.csv',delimiter=',')

# WAIT
mean = np.mean(a)

# Poisson
def w(x, mean):
    return ((mean**x) * np.exp(-mean))/(factorial(x))

x_values = np.linspace(0, 5, 1000)
y_values = w(x_values, mean)
plt.plot(x_values, y_values, color="black")

plt.hist(a, edgecolor='black', bins = 50, density = True)  # 30 bins, black edges
plt.xlabel("Years Before Phone Dropped")
plt.ylabel("Frequency")
plt.title("Histogram of Years Before Phone Dropped")
plt.show()
```

```python
from scipy.stats import expon

def ln_prob(t):
    return a.shape[0]*np.log(t) - t*np.sum(a)
def grad_ln_prob(t):
    return a.shape[0]*(1/t) - np.sum(a)

t=2
mu = 0.002

for i in range(100):
    t = t + mu*grad_ln_prob(t)


print('Using the mean method, the best θ = 1/μ = %.3f'%(1/mean))
print('Using Maximum Likelihood, the best θ = %.3f'%t)
print('Probability of drop phone within 2 years %.3f'%expon.cdf(2, 0, mean))
```
