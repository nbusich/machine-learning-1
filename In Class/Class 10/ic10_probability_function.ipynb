```python
SAT = np.genfromtxt('SAT.csv', delimiter=',')
AGE = np.genfromtxt('student_age.csv', delimiter=',')
WAIT = np.genfromtxt('lunch_wait_time.csv', delimiter=',')
```

```python
# SAT
mean = np.mean(SAT)
std = np.std(SAT)

def f(x, mean, std):
    return (1/(std*math.sqrt(2*math.pi))) * np.exp((-(x-mean)**2)/(2*std**2))

sat_pdf = f(x, mean, std)

x_values = np.linspace(700, 1600, 1000)
y_values = f(x_values, mean, std)
plt.plot(x_values, y_values, color="black")

plt.hist(SAT, edgecolor='black', density=True) 
plt.xlabel("SAT SCORE")
plt.ylabel("Frequency")
plt.title("Histogram of SAT SCORE")
plt.show()
```

```python
# AGE
mean = np.mean(AGE)
std = np.std(AGE)

def a(x):
    return np.full_like(x, 1/(23-17))
x_values = np.linspace(17, 23, 1000)
y_values = a(x_values)
plt.plot(x_values, y_values, color="black")
plt.hist(AGE, edgecolor='black', density=True)  # 30 bins, black edges
plt.xlabel("AGE")
plt.ylabel("Frequency")
plt.title("Histogram of AGE")
plt.show()
```

```python
from scipy.special import factorial
# WAIT
mean = np.mean(WAIT)

# Poisson
def w(x, mean):
    return ((mean**x) * np.exp(-mean))/(factorial(x))

x_values = np.linspace(0, 5, 1000)
y_values = w(x_values, mean)
plt.plot(x_values, y_values, color="black")

plt.hist(WAIT, edgecolor='black', density = True)  # 30 bins, black edges
plt.xlabel("WAIT TIME")
plt.ylabel("Frequency")
plt.title("Histogram of WAIT TIME")
plt.show()
```
